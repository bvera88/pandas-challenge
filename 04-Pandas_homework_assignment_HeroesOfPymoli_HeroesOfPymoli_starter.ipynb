```python
# Dependencies and Setup
import pandas as pd
import csv


# File to Load (Remember to Change These)
data_file = "Heroes_Pandas.csv"

# Read Purchasing File and store into Pandas data frame
pymoli_data = pd.read_csv(data_file)

pymoli_data.head()
```

## Player Count

* Display the total number of players


```python
Total_Number_of_Players = pymoli_data["SN"].unique()

print(len(Total_Number_of_Players))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(pymoli_data["Item ID"].unique())

average_price = pymoli_data["Price"].mean()
                             
number_of_purchase = len(pymoli_data["Purchase ID"].unique())
                             
total_revenue = pymoli_data["Price"].sum()
                             
Purchases_df = pd.DataFrame([{
    'Number of Unique Items': unique_items,
    'Average Price': average_price,
    'Number of Purchases': number_of_purchase,
    'Total Revenue': total_revenue
}])
                             
Purchases_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# im using the total count from earlier with 576 unique players
gender_demo = pd.DataFrame()


#grouping the genders here
gender_demo["Total Count"] = pymoli_data.groupby('Gender')['SN'].count()


#adding up the total count and dividing the total count
gender_demo["Percentage of Players"] = ((gender_demo["Total Count"] / pymoli_data["SN"].count()) * 100).map("{:.2f}%".format)


gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_analysis = pd.DataFrame()

#im using the SN column to add up the gender groups
gender_analysis["Purchase Count"] = pymoli_data.groupby('Gender')['SN'].count()
gender_analysis["Average Purchase Price"] = pymoli_data.groupby('Gender')['Price'].sum() / gender_analysis["Purchase Count"]
gender_analysis["Total Purchase"] = pymoli_data.groupby('Gender')['Price'].sum()

gender_analysis["Average Purchase Price"] = gender_analysis["Average Purchase Price"].map("${:.2f}".format)
gender_analysis["Total Purchase"] = gender_analysis["Total Purchase"].map("${:.2f}".format)

gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
print(pymoli_data["Age"].max())
print(pymoli_data["Age"].min())
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]

group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

print(bins)
print(group_labels)


#pandas cut method
pymoli_data["Age Group"] = pd.cut(pymoli_data["Age"], bins, labels=group_labels)

pymoli_data[pymoli_data["Age Group"]  == "40+"]
```

```python
grouped_df = pd.DataFrame()

grouped_df["Total Count"] = pymoli_data.groupby('Age Group')['SN'].nunique()

grouped_df["Percentage of Players"] = ((grouped_df["Total Count"] / len(Total_Number_of_Players)) * 100 ).map("{:.2f}%".format)

pymoli_data.groupby('Age Group')['SN'].nunique()


grouped_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# grouped_df = pd.DataFrame()

# grouped_df["Total Count"] = pymoli_data.groupby('Age Group')['SN'].nunique()

purchase_df = pd.DataFrame()

purchase_df["Purchase Count"] = pymoli_data.groupby("Age Group")["Purchase ID"].count()

purchase_df["Average Purchase Price"] = (pymoli_data.groupby("Age Group")["Price"].sum() / purchase_df["Purchase Count"]).map("${:.2f}".format)

purchase_df["Total Purchase Value"] = (pymoli_data.groupby("Age Group")["Price"].sum()).map("${:,.2f}".format)

purchase_df["Avg Total Purchase per Person"] = (pymoli_data.groupby("Age Group")["Price"].sum() / pymoli_data.groupby('Age Group')['SN'].nunique()).map("${:.2f}".format)

purchase_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = pd.DataFrame()

top_spenders["Purchase Count"] = pymoli_data.groupby("SN")["Purchase ID"].count()
top_spenders["Average Purchase Price"] = (pymoli_data.groupby("SN")["Price"].sum() / top_spenders["Purchase Count"]).map("${:,.2f}".format)
top_spenders["Total Purchase Value"] = pymoli_data.groupby("SN")["Price"].sum()

top_spenders.sort_values(by = ["Total Purchase Value"], ascending = False, inplace = True)

top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:,.2f}".format)

# .style.format("{:.2%}")

# .map("${:,.2f}".format)

top_spenders.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
cols = ["Item ID", "Item Name"]

most_popular = pd.DataFrame()


most_popular["Purchase Count"] = pymoli_data.groupby(cols)["Item Name"].count()
most_popular["Item Price"] = (pymoli_data.groupby(cols)["Price"].mean()).map("${:,.2f}".format)
most_popular["Total Purchase Value"] = (pymoli_data.groupby(cols)["Price"].sum()).map("${:,.2f}".format)

# im sorting the table in descending order, and showing top five values using .sort_values and .head(5)
most_popular.sort_values(by = ["Purchase Count"], ascending = False).head(5)

# top_spenders.sort_values(by = ["Total Purchase Value"], ascending = False, inplace = True)
```
